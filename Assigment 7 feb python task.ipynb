```python
Ans = 1
def check_password(password):
    # Password should be at least 8 characters long
    if len(password) < 8:
        return "Invalid Password"
    
    # Password should contain at least one uppercase letter, one lowercase letter, one digit and one special character
    uppercase = False
    lowercase = False
    digit = False
    special_char = False
    for char in password:
        if char.isupper():
            uppercase = True
        elif char.islower():
            lowercase = True
        elif char.isdigit():
            digit = True
        else:
            special_char = True
    
    if not (uppercase and lowercase and digit and special_char):
        return "Invalid Password"
    
    # Password is valid
    return "Valid Password"
```
